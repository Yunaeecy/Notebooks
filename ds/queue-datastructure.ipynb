# Overview

A queue is a list-like structure that provides restricted access to lts elements, it has a characteristic of "First-In, First-Out"(FIFO). Queue elements may only be inserted at the back(called an enqueue operation) and removed from the front(called a dequeue operation). Thus, queues release their elements in order of arrival. 

# Design ADT Queue
Here we are going to implement Queue in Python by using list.

Firstly, let's design our ADT Queue. Its operations include:
* is_empty
* is_full
* enqueue
* dequeue
* peek

# Implement ADT Queue

```python
class Queue:
    """
    A simple queue implementation using a list.
    """
    def __init__(self):
        self.items = [] 
        
    def is_empty(self):
        return len(self.items) == 0
    
    def enqueue(self,item):
        self.items.append(item)
            
    def dequeue(self):
        """
        Removes and returns the element from the front of the queue.
        Raises:
            IndexError: If the queue is emmpty.
        """
        if self.is_empty():
            raise IndexError("Queue is empty")
        return self.items.pop(0)
    
    def peek(self):
        """
        Returns the element at the front of the queue without removing it.
        Raises:
            IndexError: If the queue is empty.
        """
        if self.is_empty():
            raise IndexError("Queue is empty")
        return self.items[0]
    
    def size(self):
        return len(self.items)
    
q =  Queue()
```

```python
q.size()
```

```python
q.enqueue(1)
q.size()
```

```python
q.enqueue(2)
q.enqueue(3)
q.size()
```

```python
print(q.dequeue())
```

```python
print(q.peek())
```
