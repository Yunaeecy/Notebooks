```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import LabelEncoder
```

```python
# Sample data (Assuming you have already loaded your data into df)
# Replace this with your actual DataFrame containing 'Order', 'Family', and 'Genus'
df = pd.read_csv('/kaggle/input/medicinal-flora-of-australia/CMFoADB 2024_cleaned.csv')
df.head()
```

```python
df.columns
```

```python
import plotly.graph_objs as go
```

```python
# Label encoding categorical variables
le_order = LabelEncoder()
le_family = LabelEncoder()
le_genus = LabelEncoder()

df['Order_encoded'] = le_order.fit_transform(df['Order'])
df['Family_encoded'] = le_family.fit_transform(df['Family'])
df['Genus_encoded'] = le_genus.fit_transform(df['Genus'])

# Create 3D scatter plot using Plotly
fig = go.Figure(data=[go.Scatter3d(
    x=df['Order_encoded'],
    y=df['Family_encoded'],
    z=df['Genus_encoded'],
    mode='markers',
    marker=dict(
        size=5,
        color=df['Genus_encoded'],  # Color by Genus
        colorscale='Viridis',        # Color scale
        opacity=0.8
    ),
    text=df['Genus']  # Display Genus name on hover
)])

# Add axis titles
fig.update_layout(
    scene=dict(
        xaxis_title='Order',
        yaxis_title='Family',
        zaxis_title='Genus'
    ),
    margin=dict(l=0, r=0, b=0, t=0)  # Remove margin to maximize plot space
)

# Show the plot
fig.show()
```

1. 3D Plot:
3D stands for three dimensions: X-axis, Y-axis, and Z-axis.
A 3D plot allows you to visualize how three variables relate to each other in space.
Each point in a 3D scatter plot has a position based on three coordinates: (X, Y, Z).
Example: We can visualize the relationship between Order, Family, and Genus in a 3D space where each taxon is represented by a point in that space.

2. 4D Plot:
4D stands for four dimensions, which goes beyond the three physical axes (X, Y, Z).

**Since we can't add another spatial axis beyond the three dimensions in physical space, we represent the fourth dimension through:**
Color: Each points color can represent a different variable, such as Species in our example. Different colors indicate different values or categories.
Size: Another option is to represent the fourth dimension with the size of each point, where larger points correspond to higher values in that additional dimension.
Opacity: Sometimes, transparency (or opacity) of the points is used to represent the fourth dimension.

Difference in Practice:

**3D Plot:**
You see three variables (e.g., Order, Family, Genus) positioned in space.
Each point is fixed at a position determined by the values of those three variables.

**4D Plot:**
You still visualize the three dimensions (X, Y, Z) as before, but a fourth dimension (e.g., Species) is added.
The fourth dimension can be visualized via color, size, or other visual attributes like opacity. For instance, in a 4D plot:
The X, Y, and Z axes represent Order, Family, and Genus.
The color (or size) of the points represents the fourth variable (Species).
Example of a 4D Plot:
In a 4D scatter plot of taxonomic data:

X-axis: Represents Order.
Y-axis: Represents Family.
Z-axis: Represents Genus.
Color (4th dimension): Could represent Species. Each points color varies based on the species it belongs to.

Visual Summary:
Dimension	Representation in 3D Plot	Representation in 4D Plot
X-Axis	Represents a variable (e.g., Order)	Represents a variable (e.g., Order)
Y-Axis	Represents a variable (e.g., Family)	Represents a variable (e.g., Family)
Z-Axis	Represents a variable (e.g., Genus)	Represents a variable (e.g., Genus)
4th Dimension	N/A	Represented by color or size (e.g., Species)

In Summary:
3D plots allow you to visualize three variables in a physical space.
4D plots visualize a fourth variable using visual encodings like color or size, giving more information in the same plot.

Both 3D and 4D plots are useful for visualizing relationships between multiple variables, with 4D allowing you to represent an additional layer of data.


```python
# Check if 'Species' has missing values
print(f"Missing values in 'Species': {df['Species'].isnull().sum()}")

# Handle missing values by filling with 'Unknown' or dropping rows with NaN in 'Species'
df['Species'].fillna('Unknown', inplace=True)

# Label encoding categorical variables
le_order = LabelEncoder()
le_family = LabelEncoder()
le_genus = LabelEncoder()
le_species = LabelEncoder()

# Encode the columns
df['Order_encoded'] = le_order.fit_transform(df['Order'])
df['Family_encoded'] = le_family.fit_transform(df['Family'])
df['Genus_encoded'] = le_genus.fit_transform(df['Genus'])
df['Species_encoded'] = le_species.fit_transform(df['Species'])  # Encode 'Species' column

# Create 4D scatter plot using Plotly
fig = go.Figure(data=[go.Scatter3d(
    x=df['Order_encoded'],
    y=df['Family_encoded'],
    z=df['Genus_encoded'],
    mode='markers',
    marker=dict(
        size=df['Species_encoded'] / 2,  # Vary size by the Species encoded value
        color=df['Species_encoded'],  # Color by Species (4th dimension)
        colorscale='Viridis',         # Color scale
        opacity=0.8
    ),
    text=df['Species']  # Display Species name on hover
)])

# Add axis titles
fig.update_layout(
    scene=dict(
        xaxis_title='Order',
        yaxis_title='Family',
        zaxis_title='Genus'
    ),
    margin=dict(l=0, r=0, b=0, t=0)  # Remove margin to maximize plot space
)

# Show the plot
fig.show()
```

### Explanation:

* Label Encoding: We encode the categorical variables using LabelEncoder for Order, Family, Genus, and Species to transform them into numeric values.

* Plotly 4D Visualization: We create a 3D scatter plot and use color and size to represent the encoded Species values (the fourth dimension).

* Expected Outcome:
We should see a 4D scatter plot where:

X-Axis: Represents Order.
Y-Axis: Represents Family.
Z-Axis: Represents Genus.
Color and Size: Represents the Species column.



From the 4D plot, we can interpret:

1. Three-Dimensional Taxonomic Relationships (X, Y, Z axes):
The plot shows the relationships between Order (X-axis), Family (Y-axis), and Genus (Z-axis) for the taxa in your dataset.
The spread of the points across the axes can give insight into how diverse the genera are within each order and family. You can observe whether certain families tend to have more genera clustered within specific orders.

2. Color Coding (4th Dimension: Species):

Color & Size: Both reflect Species_encoded:
Color: Indicates which species a point corresponds to.

Size: Also reflects the encoded species, providing an additional way to differentiate species visually.


Insight: You can see how different species are distributed across the genus, family, and order. If certain colors (species) are clustered around certain areas of the plot, it means that those species are closely related taxonomically.

3. Point Size (Species Encoding):
The size of the points is based on the Species_encoded value. Larger points represent larger Species_encoded values, while smaller points represent smaller values.
Insight: The variation in point size can help identify which species are more prominent or frequent in the dataset, though it might also indicate an artifact of encoding.

4. Overlapping Points:
There are many overlapping points, which can indicate that there are many species within a particular genus or family. This means some genera might contain multiple species that are closely related.

Insight: A dense cluster of large, overlapping points suggests that many species are within the same taxonomic hierarchy.

What Can We Learn:

**Taxonomic Hierarchy**: The plot helps us understand the hierarchical distribution of species within genera, families, and orders. We can observe how diverse these taxa are in the context of the dataset.

**Species Distribution**: The color variation provides insight into how species are distributed across the taxonomic levels. Dense clusters in certain areas might indicate high species diversity within those particular genera or families.

**Species Prevalence**: Larger points suggest certain species are more prevalent or significant in the data. This could indicate either a higher number of occurrences or a higher weight given to them during encoding.
