# Bubble Sort: A gentle introduction:

Bubble sort is a simple sorting algorithm that repeatly steps through the list, compares adjacent elements and swaps them if they are in the wrong order. This process is repeated until the list is sorted.

# Characteristics:

`Simple to understand and implement`:The logic behind bubble sort is straightward, making it a good choice for educational purposes.

`Inefficient for large lists`: Bubble sort has a time complexity of O(n^2), meaning that the time taken to sort grows quadratically with the number of elements. This makes it inefficient for large datasets.

`In-place sorting`: It doesn't require additional memory space to sort the elements.

# Bubble Sort Time Complexity: Best, Worst, and Average Cases

As you mentioned, the time complexity of bubble sort varies depending on the initial order of the elements. Let's explore each case:

1. `Worst-Case Scenario: O(n^2)`

This occurs when the array is sorted in reverse order.
In this case, every element needs to be compared and swapped with its neighbor on every pass.

The number of passes required is (n-1), and within each pass, we perform (n-1) comparisons and potential swaps.

Therefore, the total number of operations is proportional to n * (n-1), leading to a time complexity of O(n^2).

2. `Best-Case Scenario: O(n)`

This occurs when the array is already sorted.

In this case, the first pass through the array will identify that no swaps are needed as all elements are in their correct positions.

The algorithm will recognize the sorted order and terminate after just one pass, resulting in a linear time complexity of O(n).

3. `Average-Case Scenario: O(n^2)`

On average, assuming a random distribution of elements, bubble sort will still require a quadratic number of operations.

While some elements might be in their correct positions, a significant number of comparisons and swaps will be needed to fully sort the array.

The average-case time complexity remains O(n^2), similar to the worst-case scenario.

**In summary:**

Worst-case:` O(n^2)` (reverse-sorted array)

Best-case: `O(n)` (already sorted array)

Average-case: `O(n^2)` (randomly ordered array)

```python
def bubble_sort(data):
    n=len(data)
    for i in range(n-1): #Iterate through that list (n-1) times
        for j in range(n-i-1): #Compare and swap elements
            if data[j]>data[j+1]:
                data[j],data[j+1]=data[j+1],data[j]
# Example list
numbers=[5,3,8,6,7,2]
bubble_sort(numbers)
print(numbers)
```
