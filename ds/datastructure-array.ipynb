```python
# array
class PartyArray:
    def __init__(self):
        self.data = []
    def __len__(self):
        return len(self.data)
    def remove(self,value):
        self.data.remove(value)
    def append(self, value):
        """
        add new value to the end
        """
        self.data.append(value)

p=PartyArray()
print(type(p))
```

```python
p.append(1)
print(p)
```

```python
print(len(p))
```
