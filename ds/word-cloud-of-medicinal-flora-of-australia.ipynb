```python
import pandas as pd

# Change 'your_file.csv' to the actual path
df = pd.read_csv('/kaggle/input/medicinal-flora-of-australia/CMFoADB 2024_cleaned.csv')

# Check the first few rows of the data
df.head()
```

<div style="background-color: pink; color: black; padding: 10px; border-radius: 10px; font-size: 24px;">
  Word Cloud for the Species Column
</div>


```python
# Generate word cloud from the 'Species' column
text = " ".join(df['Species'].astype(str))

# Create and display word cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
