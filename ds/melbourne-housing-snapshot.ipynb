```python
plt.figure(figsize=(10, 8))
sns.barplot(data= df.groupby('Type')['Price'].mean().reset_index(), x='Type', y='Price', palette='coolwarm')
plt.title('Average Price by Type')
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```
