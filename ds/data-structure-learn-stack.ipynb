# Overview

In this notebook, we are going to implement Stack data structure in Python. A stack is a ***linear data structure*** that operates based on the **Last-in-First-out(LIFO)** priciple. This is the most fundamental and defining characteristic of it. A stack defines the operations that can be performed o the data, but not the specific iplementation details(ADT).

The common operations on a stack.
* push(item) : Adds an element(item) to the top of the stack.
* pop(): Removes and returns the top element from the stack.
* peek(): Returns the top element without removing it.
* isEmpty(): Checks if the stack is empty and returns a boolean value(True/False).
* size(): Returns the number of elements currently in the stack.

# Define Stack(ADT)

```python
class Stack:
    def __init__(self):
        self.items = []
    def push(self,item):
        self.items.append(item)
    def pop(self):
        if self.isEmpty():
            raise Exception("Stack is empty")
        else:
            return self.items.pop()
    def isEmpty(self):
        """
        checks if the stack is empty.
        Returns:
            bool: True if the stack is empty, otherwise False.
        """
        return self.items == []
    def peek(self):
        if self.isEmpty():
            raise Exception("Stack is empty")
        else:
            return self.items[-1]
    def size(self):
        return len(self.items)

myStack =Stack()
```
