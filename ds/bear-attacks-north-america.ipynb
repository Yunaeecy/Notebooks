```python
# Handling null values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
bear_attacks['Age']=imputer.fit_transform(bear_attacks[['Age']])
bear_attacks['Latitude']=imputer.fit_transform(bear_attacks[['Latitude']])
bear_attacks['Longitude']=imputer.fit_transform(bear_attacks[['Longitude']])
imputer2 = SimpleImputer(strategy='most_frequent')
bear_attacks['gender']=imputer2.fit_transform(bear_attacks[['gender']]).ravel()
```

```python
bear_attacks.isna().sum()
```

```python
# Convert column data types to appropriate formats
bear_attacks['Age'] = bear_attacks['Age'].round().astype('Int64')
bear_attacks['Date'] = pd.to_datetime(bear_attacks['Date'],errors='coerce')
bear_attacks['Type of bear'] = bear_attacks['Type of bear'].astype('category')
bear_attacks['Hunter'] = bear_attacks['Hunter'].astype('bool')
bear_attacks['Grizzly'] = bear_attacks['Grizzly'].astype('bool')
bear_attacks['Hikers'] = bear_attacks['Hikers'].astype('bool')
bear_attacks['Only one killed'] = bear_attacks['Only one killed'].astype('bool')
```

```python
# Collect numeric columns into list
columns=bear_attacks.select_dtypes(include=['number']).columns.tolist()
columns
```

```python
# handling outliers
for col in columns:
        
    Q1 = bear_attacks[col].quantile(0.25)
    Q3 = bear_attacks[col].quantile(0.75)
    IQR = Q3 - Q1

    outliers = ((bear_attacks[col] < (Q1 - 1.5 * IQR)) | (bear_attacks[col] > (Q3 + 1.5 * IQR)))
    print(outliers.value_counts())
    print('*' * 20)
```

```python
bear_attacks_edited=bear_attacks.drop(columns=['Location','Description'])
```

```python
bear_attacks_edited.head(5)
```

```python
# Encoding categorical variables
bear_attacks_edited = pd.get_dummies(bear_attacks_edited, columns=['gender', 'Type of bear'])
```

```python
bear_attacks_edited.head(5)
```

```python
print(bear_attacks_edited.columns)
```

```python
bear_attacks_edited.columns = bear_attacks_edited.columns.str.strip()
```

```python
print(bear_attacks_edited.columns.tolist())
```

```python
X = bear_attacks_edited[['Age', 
                          'gender_female', 'gender_male', 
                          'Hunter', 'Grizzly', 'Hikers', 
                          'Type of bear_Brown bear', 
                          'Type of bear_Polar Bear']]

y = bear_attacks_edited['Only one killed']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# Initializing and fitting the model
model = LogisticRegression()
model.fit(X_train , y_train)
```

```python
# Making predictions
y_pre=model.predict(X_test)
y_pre
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test,y_pre)
accuracy
```

```python
confusion_matrix(y_test , y_pre)
```
