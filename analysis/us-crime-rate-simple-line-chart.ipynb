```python
import plotly.express as px
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/kaggle/input/us-crime-rates-1960-2014/US_Crime_Rates_1960_2014.csv")
df.head()
```

```python
df.info()
```

Now, we convert the 'Year' column to a datetime format, assuming the 'Year' column contains date information in the format '%Y' (which represents the year with century as a decimal number).

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
```

```python
# count the number of duplicate rows in a pandas DataFrame df. 
#The duplicated() function identifies duplicate rows, 
#and sum() counts the number of True values (i.e., the number of duplicate rows).
df.duplicated().sum()
```

```python
# generate descriptive statistics of a DataFrame
df.describe()
```

```python
sns.set_style("darkgrid",{"axes.facecolor":"#282828","grid.color":"#282828","text.color":"white"})
plt.figure(figsize=(10,6))

# Seaborn line plot banane ke liye
sns.lineplot(x='Year',y='Murder',data=df,marker='o',color='purple',label="Murder")
sns.lineplot(x='Year',y='Forcible_Rape',data=df,marker='o',color='orange',label='Forcible Rape')
sns.lineplot(x='Year',y='Robbery',data=df,marker='o',color='green',label='Robbery')
sns.lineplot(x='Year',y='Aggravated_assault',data=df,marker='o',color='blue',label='Aggravated Assault')
sns.lineplot(x='Year',y='Larceny_Theft',data=df,marker='o',color='pink',label='Larceny_Theft')
sns.lineplot(x='Year',y='Vehicle_Theft',data=df,marker='o',color='pink',label='Vehicle_Theft')
sns.lineplot(x='Year',y='Property',data=df,marker='o',color='yellow',label='Property')

# Plot katitle aur label set karna
plt.title('Crime Data Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')

# Legend ko display karna
plt.legend()

# Plot dikhane ke liye
plt.show()
```
