# Area chart

An area chart is a type of graph that is used to show `how a quantity changes over time` . It is similar to a line chart, but the area below the line is filled in with color or shading. This can make it easier to see trends and how different quantities are related.

**Area charts are often used to show:

1.Trends over time, such as sales figures or website traffic

2.How different parts of a whole make up the total, such as the breakdown of a company's revenue by product

3.How two or more quantities are related, such as the relationship between temperature and rainfall


**Area charts can be a good choice when you want to:

1.Show how a quantity has changed over time

2.Emphasize the magnitude of a quantity

3.Compare two or more quantities


```python
import numpy as np
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
```

# Declaring data

```python
# Create some example data
python=np.array([2, 3, 7, 5, 26, 221, 44, 233, 254, 265, 266, 267, 120, 111])
pypy=np.array([12, 33, 47, 15, 126, 121, 144, 233, 254, 225, 226, 267, 110, 130])
jython=np.array([22, 43, 10, 25, 26, 101, 114, 203, 194, 215, 201, 227, 139, 160])

dims = dict(kdims='time', vdims='memory')
python = hv.Area(python, label='python', **dims)
pypy   = hv.Area(pypy, label='pypy', **dims)
jython = hv.Area(jython, label='jython', **dims)
```

# Plot

```python
opts.defaults(opts.Area(fill_alpha=0.5))
overlay = (python * pypy * jython)
overlay.relabel("Area Chart") + hv.Area.stack(overlay).relabel("Stacked Area Chart")
```

# Example

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with the actual filename)
df = pd.read_csv('/kaggle/input/retail-e-commerce-sales-inactive/20100072.csv')

# Print the column names
print(df.columns)

# Print the first five rows
print(df.head())

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Convert the 'REF_DATE' column to datetime
df['REF_DATE'] = pd.to_datetime(df['REF_DATE'])

# Sort the DataFrame by date
df = df.sort_values(by='REF_DATE')

# Plotting the area chart
plt.figure(figsize=(10, 6))
plt.fill_between(df['REF_DATE'], df['Sales'], color='skyblue', alpha=0.4, label='Sales')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Over Time')

# Show the plot
plt.legend()
plt.show()
```

The purpose of generating an area chart using the provided code is to visually represent the trend and distribution of sales over time. Area charts are effective for illustrating cumulative values and variations in a dataset.
