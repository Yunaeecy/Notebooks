# Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
%matplotlib inline
warnings.filterwarnings('ignore')
```

# EDA

```python
df = pd.read_csv('/kaggle/input/black-friday/train.csv')
df
```

```python
df.info()
```

```python
df.describe().round(2)
```

```python
df.drop(['User_ID'], axis = 1, inplace = True)
```

```python
df.nunique()
```

```python
df['Gender'] = df['Gender'].map({'F':0,'M':1})
```

```python
df['Age'].unique()
```

```python
df['Age'] = df['Age'].map({'0-17': 0, '18-25' : 1,'26-35':2,'35-45':3,'46-50':4,'51-55':5,'55+':6})
#df['Age'] = df['Age'].map({'0-17':0, '18-25':1, '26-35':2, '36-45':3, '46-50':4, '51-55':5, '55+':6})
df
```

```python
df['City_Category'].unique()
```

```python
df['City_Category'] = df['City_Category'].map({'A':0,'B':1,'C':2})
df
```

```python
df.drop(['Product_ID'], axis = 1, inplace = True)
df
```

```python
df['Product_Category_2'].unique()
```

```python
#filling the missing values in the 'Product_Category_2' column with the mode (most frequent value) of that column. 
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
df
```

```python
#count the number of missing values (NaN) in each column of the DataFrame df
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
df.isnull().sum()
```

```python
#remove the '+' character from the values in the 'Stay_In_Current_City_Years' column
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.sample(10)
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['Product_Category_2'] = df['Product_Category_2'].astype(int)
df['Product_Category_3'] = df['Product_Category_3'].astype(int)
```

```python
df
```

```python
df.info()
```

# Modeling

```python
y = df['Purchase'] 
X = df.drop(['Purchase'],axis=1)
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
sc = StandardScaler()
sc.fit_transform(X_train)
```

```python
sc.fit_transform(X_test)
```

```python
lm=LinearRegression()
lm.fit(X_train, y_train)
w=lm.coef_
b=lm.intercept_
print(w)
print(b)
```

```python
predictions=lm.predict(X_test)
predictions
```

```python
RF = RandomForestRegressor(n_estimators=10,max_depth=7,n_jobs=-1,random_state=42)
RF.fit(X_train,y_train)
```

```python
y_pred = RF.predict(X_test)
y_pred
```

```python
y_test
```

```python
MAE=mean_absolute_error(y_test,y_pred)
MAE
```

```python
r2_score(y_test,y_pred)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
y.mean()
```

```python
y_test.mean()
```

```python
xgb_reg=XGBRegressor(n_estimators=480, learning_rate=0.2, max_depth=8)
```

```python
xgb_reg.fit(X_train,y_train)
```

```python
xgb_y_pred=xgb_reg.predict(X_test)
```

```python
print("RMSE:",np.sqrt(mean_squared_error(y_test, xgb_y_pred)))
print("y_mean",(y.mean()))
print("R2 Score:",round(r2_score(y_test,xgb_y_pred)*100,2))
```
