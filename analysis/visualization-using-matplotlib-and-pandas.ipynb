```python
iris = pd.read_csv('/kaggle/input/ipl-ball-by-ball-batsman-season-batter-iris/iris.csv')
iris
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],c=iris['Species'], cmap='summer')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],c=iris['Species'], cmap='plasma')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],c=iris['Species'], cmap='plasma', alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

The `alpha` parameter is used to control the **transparency** of the points in the scatter plot.

Setting **alpha=0.7** means that each data point plotted will have an opacity of 70%.

An `alpha value of 1.0` would mean **fully opaque (no transparency)**, while a `value of 0.0` would mean **fully transparent (invisible)**.

In this case, an alpha value of 0.7 adds some transparency to the points, allowing some degree of overlap to be visible if multiple points are plotted in the same location. This can be helpful in visualizing the density of points in areas with high data concentration.

# Annotations
Point Name, size and Color `plt.text(1,5, 'Point 1', fontdict={'size':12, 'color':'brown'})`

```python
batter = pd.read_csv('/kaggle/input/ipl-data-set/batter.csv')
batter.head()
```

```python
sample_df = batter.head(100).sample(25, random_state=5)
sample_df
```

```python
plt.figure(figsize=(18, 10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

# Horizontal and Vertical lines on graph

```python
plt.figure(figsize=(18, 10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

plt.axhline(120, color='red')
plt.axhline(140, color='blue')
plt.axvline(25, color='green')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

# Subplots

```python
# A different way to plot graphs
batter.head()
```

```python
# 1st way
plt.figure(figsize=(12, 6))

plt.scatter(batter['avg'], batter['strike_rate'])
plt.title('Average vs Strike_rate')
plt.xlabel('Average')
plt.ylabel('Strike_rate')

plt.show()
```

```python
# 2nd way
fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(batter['avg'], batter['strike_rate'], color='red', marker='+')
ax.set_title('Average vs Strike_rate')
ax.set_xlabel('Average')
ax.set_ylabel('Strike_Rate')

fig.show()
```

```python
# (fig) 1st object and (ax) is the axis object
fig, ax = plt.subplots()

# passing the values of (axis) in ax
ax.plot(batter['avg'], batter['strike_rate'], marker='o')
ax.set_title('Average vs Strike_rate')
ax.set_xlabel('Average')
ax.set_ylabel('Strike_Rate')
```

```python
# subplots
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(12, 6))

ax[0].scatter(batter['avg'], batter['strike_rate'], color='red')
ax[1].scatter(batter['avg'], batter['runs'])

ax[0].set_title('Avg vs Strike_Rate')
ax[0].set_ylabel('Strike_Rate')

ax[1].set_title('Avg vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# subplots scatter and histogram
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax[0, 0].scatter(batter['avg'], batter['strike_rate'], color='red')
ax[0, 1].scatter(batter['avg'], batter['runs'])
ax[1, 0].hist(batter['avg'])
ax[1, 1].hist(batter['runs'])
```

```python
# 3rd way
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'], batter['strike_rate'], color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batter['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'])
```
