```python
import plotly.express as px
```

```python
df = px.data.stocks()
fig = px.line(df, x='date', y='GOOG', title='Interactive Line Chart')
fig.show()
```

```python
df = px.data.iris()
fig = px.scatter(df, x='sepal_width', y='sepal_length', color='species', size='petal_length',
                 title='Dynamic Scatter Plot')
fig.show()
```
