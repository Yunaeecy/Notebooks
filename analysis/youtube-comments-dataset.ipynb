# Step 1: Load the Dataset and Explore

```python
import pandas as pd
import re
```

```python
# Load the dataset
df = pd.read_csv('/kaggle/input/youtube-comments-dataset/YoutubeCommentsDataSet.csv')

# Check the first few rows
df.head()
```

```python
# Check for Missing Values
df.isnull().sum()
```

```python
# Check Data Types
df.info()
```

# Step 2: Improve the Dataset

```python
# Standardize Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
#Remove Duplicates
df = df.drop_duplicates()
```

```python
print(df.columns)
```

```python
# Clean Text (Remove Special Characters, Links, etc.)

def clean_text(text):
    text = str(text).lower()  # Convert to lowercase
    text = re.sub(r'http\S+', '', text)  # Remove links
    text = re.sub(r'@\w+', '', text)  # Remove mentions
    text = re.sub(r'#\w+', '', text)  # Remove hashtags
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove special characters
    return text.strip()
```

```python
# Apply text cleaning and additional features
df['cleaned_comment'] = df['comment'].apply(clean_text)
df['word_count'] = df['comment'].apply(lambda x: len(str(x).split()))
df['char_count'] = df['comment'].apply(lambda x: len(str(x)))
```

# Step 3: Save the Cleaned Dataset

```python
# Save cleaned dataset in Kaggle Notebook
output_path = "/kaggle/working/YoutubeCommentsDataSet_Cleaned.csv"
```

```python
df.to_csv(output_path, index=False)
```

```python
print("✅ Dataset cleaned and saved successfully at:", output_path)
```
